```python
#Question 12: Retrieving State code from Municipality code
def codeProvince(codeCommune):
    return codeCommune[:3] if len(codeCommune) > 3 else None
```

```python
#Question 12: Adding State code column
dfc.insert(3, "Code_Province", dfc["Code_Commune"].apply(codeProvince))
```

```python
#Question 13: Retrieving State name by State code
def nomProvince(codeProvince):
    try:
        df = pd.read_csv("data_python/code_nom_province.csv", dtype=str)
        df.set_index("Code_Province", inplace=True)
        return df.loc[codeProvince][0]
    except:
        return None
```

```python
#Question 13: Adding State name column
dfc.insert(4, "Nom_Province", dfc["Code_Province"].apply(nomProvince))
```

```python
#Question 14: Setting State and Municipality as indexes
dfc.set_index(["Code_Province", "Code_Commune"], inplace=True)
dfc.sort_index(inplace=True)
```

```python
#Question 14: Exporting DataFrame
dfc.to_excel("data_python/dfc_Abidar_El_Mehdi.xlsx")
```

```python
#Question 16: Calculating number of municipality by type for every State
dfc.groupby(["Code_Province"])["Type_Commune"].value_counts()
```

```python
#Question 17: Calculating activity rate
dfc.groupby("Code_Province").apply(lambda a: a["Pop_1560_pc"].sum()/a["Population"].sum())
```

```python
#Question 18: Generating dpf Dataframe
dfp = dfc.groupby(["Nom_Province", "Code_Province"])[["Surface_ha", "Population", "Pop_1560_pc"]].sum()
dfp.rename(columns={"Pop_1560_pc": "Population_active"}, inplace=True)
dfp
```

```python
#Question 19: Generating dpf Dataframe
dfp["Taux_activite"] = dfp["Population_active"] / dfp["Population"]
dfp["Taux_analphabetisme"] = dfc.groupby(["Nom_Province", "Code_Province"])["Taux_Analphabetisme"].sum()
dfp
```

```python
#Question 20: Adding Proportions by age
dfp["Pop_inf6_pc"] = dfc.groupby(["Nom_Province", "Code_Province"])["Pop_inf6_pc"].sum()
dfp["Pop_615_pc"] = dfc.groupby(["Nom_Province", "Code_Province"])["Pop_615_pc"].sum()
dfp["Pop_1560_pc"] = dfc.groupby(["Nom_Province", "Code_Province"])["Pop_1560_pc"].sum()
dfp["Pop_sup60_pc"] = dfc.groupby(["Nom_Province", "Code_Province"])["Pop_sup60_pc"].sum()
dfp
```

```python
import requests
def get_province_data_url(province_code):
    # Themes :
    # 2 for demographic theme (Population & First mariage age)
    # 4 for education theme theme (Illiteracy rate)
    # 6 for Activity/Jobs theme (Activity rate)
    HEADERS = {
        'Referer': 'http://rgphentableaux.hcp.ma/Default1/'
    }
    THEMES = [2, 4, 5]

    data = {"Code_commune": province_code}

    for theme in THEMES:
        url = f"http://rgphentableaux.hcp.ma/Default1/getDATA/?type=Commune&CGEO={province_code}.&them={theme}"
        result = requests.get(url, headers=HEADERS).json()
        for line in result:
            indicator = line.get('INDICATEUR')

            if indicator == 'IEE_Population municipale':
                data['population'] = line["DATA2014"]

            if indicator == 'IRE_Âge moyen au premier mariage':
                data['age_moyen_mariage'] = line["DATA2014"]

            if indicator == "IEE_Taux d'analphabétisme":
                data['taux_analphabetisme'] = line["DATA2014"]

            if indicator == "IEE_Taux net d'activité":
                data['taux_activite'] = line["DATA2014"]

    return data
```
