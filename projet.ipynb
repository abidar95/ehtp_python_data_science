```python
#Question 24: Function to retrieve data from rgph site by state code
import requests
def get_commune_data_url(code_commune):
    # Themes :
    # 2 for demographic theme (Population & First mariage age)
    # 4 for education theme theme (Illiteracy rate)
    # 6 for Activity/Jobs theme (Activity rate)
    HEADERS = {
        'Referer': 'http://rgphentableaux.hcp.ma/Default1/'
    }
    THEMES = [2, 4, 5]

    data = {'Code_Commune': code_commune}

    for theme in THEMES:
        url = f'http://rgphentableaux.hcp.ma/Default1/getDATA/?type=Commune&CGEO={code_commune}.&them={theme}'
        result = requests.get(url, headers=HEADERS).json()
        for line in result:
            indicator = line.get('INDICATEUR')

            if indicator == 'IEE_Population municipale':
                data['Population_2014'] = int(line['DATA2014'])

            if indicator == 'IUE_Âge moyen au premier mariage' and line['DATA2014'] != '-':
                data['Age_moyen_mariage_2014'] = line['DATA2014']
            elif indicator == 'IRE_Âge moyen au premier mariage':
                data['Age_moyen_mariage_2014'] = line['DATA2014']

            if indicator == "IEE_Taux d'analphabétisme":
                data['Taux_analphabetisme_2014'] = line['DATA2014']

            if indicator == "IEE_Taux net d'activité":
                data['Taux_activite_2014'] = line['DATA2014']

    return data
```

```python
#Question 24: Retrieving data from rgph site by state code
# dfc.reset_index(inplace=True)
dfc_from_url = pd.DataFrame([get_commune_data_url(code_commune) for code_commune in dfc['Code_Commune']])
# dfc2 = pd.merge(dfc, dfc_from_url, on='New_Code_Commune')
dfc_from_url.head()
```

```python
# dfc_from_url.drop_duplicates('Code_Commune')['Code_Commune'].describe()
dfc2 = pd.merge(dfc.drop_duplicates('Code_Commune'), dfc_from_url.drop_duplicates('Code_Commune'), on='Code_Commune')
dfc2.head()
```

```python
#Question 26: Calculating Population growth rate by Commune
dfc2['Taux_Acroissement'] = ((dfc2['Population_2014'] - dfc2['Population']) / dfc2['Population']).round(2)
dfc2['Taux_Acroissement'].head()
```

```python
#Question 27: Calculating Population growth rate by State
dfc2.groupby(['Code_Province', 'Nom_Province'])['Taux_Acroissement'].mean().round(2)
```

```python
#Question 28: Variation of population by state between 2004 and 2014
dfc_for_graph = dfc2.groupby('Nom_Province')['Population','Population_2014'].sum()
dfc_for_graph.plot(y=['Population','Population_2014']).set(
    title='Population 2004|Population 2014',
    xlabel='State',
    ylabel='Population 2004|Population 2014'
)
```
